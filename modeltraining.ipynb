## Importing Important Libraries

```python
import pandas as pd
import numpy as np
import os
import shutil
import cv2 
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

## Assesing Dataset

```python
base_dir = "E:\procom24\Round1_B"
train_images = os.path.join(base_dir,"train_images")
train_dir = os.path.join(base_dir,"train")
test_dir = os.path.join(base_dir,"test_images")
```

```python
print(os.listdir(train_images))
```

## Preprocessing Images & matching target variables

```python
#os.mkdir(train_dir)
#os.mkdir(test_dir)
df = pd.DataFrame()
temp = []

for i in traindf['ID']:

    # print(i)
    
    imagepath = os.path.join(train_images, i)

    original_image = cv2.imread(imagepath)

    # Check if the image is successfully loaded
    if original_image is None:
        print("Error: Could not open or find the image.")
    else:
        # Convert the image to grayscale
        grayscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

        temp.append(grayscale_image.ravel())
        #temp.append(traindf['Label'][j])

        # Wait for a key event and close the windows

        cv2.waitKey(0)
        cv2.destroyAllWindows()
    
array = np.array(temp)
imageonlydf =pd.DataFrame(array)

```

```python
imageonlydf['label'] = traindf['Label']
```

```python
imageonlydf
```

```python
X = imageonlydf.drop('label', axis = 1, inplace = False)
```

```python
mmscaler.fit(X)
```

```python
X = mmscaler.transform(X)
```

```python
prediction = {}
for test_img in os.listdir(test_dir):
    imagepath = os.path.join(test_dir, test_img)
    original_image = cv2.imread(imagepath)
    if original_image is None:
        print("Error: Could not open or find the image.")
    else:
        grayscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY).ravel()
    
    if len(grayscale_image) != 3145728:
        grayscale_image = np.resize(grayscale_image,(3145728,))
        #print(grayscale_image.shape)
    
    data = kbest.transform([grayscale_image])
    prediction[test_img] = model3.predict(data)
```
